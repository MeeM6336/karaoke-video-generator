```python
import whisperx
import requests
import json
import re
from whisperx.utils import get_writer
```

```python
device = "cuda"
batch_size = 16
compute_type = "float16"
```

```python
model = whisperx.load_model("large-v3", device, compute_type=compute_type)
audio = whisperx.load_audio("output/htdemucs_ft/ayby/vocals.wav")
result = model.transcribe(audio, batch_size=batch_size)
```

```python
# Filter transcription segments with low confidence
filtered_segments = [segment for segment in result["segments"] if segment["avg_logprob"] > -0.5]

filtered_segments
```

```python
model_a, metadata = whisperx.load_align_model(language_code="en", device=device)
result_whisperx = whisperx.align(filtered_segments, model_a, metadata, audio, device, return_char_alignments=False)
```
